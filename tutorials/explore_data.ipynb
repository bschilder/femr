```python
import piton.datasets
import json
import matplotlib.pyplot as plt
import numpy as np


data = piton.datasets.PatientDatabase("/local-scratch/nigam/projects/ethanid/piton_1_extract")
```

```python
print(data.get_code_dictionary().index("Visit/IP"))
```

```python

count_data = json.load(open("/local-scratch/nigam/projects/ethanid/piton/native/results/final_counts"))

import bisect
print(len(data))

print(len(count_data))

all_counts = sorted(list(count_data))

lengths, counts = list(zip(*all_counts))

total_counts = np.cumsum(counts)
print(total_counts)

total_counts = 1 - total_counts / total_counts[-1]

plt.xscale('log')
plt.plot(lengths, total_counts)
plt.xlabel('Sequence length')
plt.ylabel('Fraction of patient sequences longer than x')

print(lengths[bisect.bisect_left(1 - total_counts, 0.5)])
print(lengths[bisect.bisect_left(1 - total_counts, 0.9)])


pos = bisect.bisect_left(lengths, 512)
print(lengths[pos], total_counts[pos])
pos = bisect.bisect_left(lengths, 1024 * 2)
print(lengths[pos], total_counts[pos])
```
